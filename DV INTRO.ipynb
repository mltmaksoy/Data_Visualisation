## Matplotlib 

### Line

```python
# Data olusturma

x= np.linspace(0, 10, 20)

y = 2 * x
```

```python
# grafik olusturma

plt.figure(figsize = (8, 6))
plt.plot(x, y)

plt.show() # sadece Grafigi göster
```

```python
# grafik olusturma

plt.figure(figsize = (5, 3))
plt.plot(x, y);  # sadece Grafigi göster
```

### Bar Plot

```python
x = np.array(["A", "B", "C", "D", "E"])
y = np.array([5, 11, 3, 14, 8])
```

```python
plt.bar(x,y)
plt.show()
```

### Histogram

```python
x = np.random.normal(170, 10, 250)  # 170:mean, 10:std, 250: sayi(gözlem)
```

```python
plt.hist(x)
plt.show()
```

### Box Plot

```python
x = np.random.normal(170, 10, 250)  # 170:mean, 10:std, 250: sayi(gözlem)
```

```python
plt.figure(figsize = (10, 6))
plt.boxplot(x);
```

```python
x = np.random.normal(170, 10, 250)  # 170:mean, 10:std, 250: sayi(gözlem)

plt.figure(figsize = (10, 6))
plt.boxplot(x);                  # Random rastgele degerler atadigi icin her seferinde farkli sonuc veriyor.
```

### Pie Chart

```python
x = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4']
```

```python
# Veri
sizes = [15, 30, 45, 10]

# Pasta grafiğini oluştur
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=x)

# Göster
plt.axis('equal')
plt.show()
```
